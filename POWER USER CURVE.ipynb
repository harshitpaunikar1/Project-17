### From here I am rearranging and manupulating the columns of the dataset ( note: no changes are made to the original dataset)

```python
# transposing dataset
b=a.transpose()
# rearranging the index of dataset
c=b.set_index([pd.Index(range(1,32))])
x=c
# renaming the index name
c.index.name = 'no_of_days ===>'
# transposing the index accordingly
y=x.T 
# renaming the column names accordingly ( so that we can understand easily )
#y.columns = ["day_1","day_2","day_3","day_4","day_5","day_6","day_7","day_8","day_9","day_10","day_11","day_12","day_13","day_14","day_15","day_16","day_17","day_18","day_19","day_20","day_21","day_22","day_23","day_24","day_25","day_26","day_27","day_28","day_29","day_30","day_31"]
y1=y
y1.head()
```

```python
# counting all unique ids present in the dataset
valuecount=pd.DataFrame(y1.stack().value_counts())

# creating separate columns "no_of_days_peopleuse_app" consistiong of no. of uccurance of the ids in dataset
valuecount.columns=["no_of_days_peopleuse_app"]

# resetting the index for convinience
valuecount.reset_index(level=0, inplace=True)
valuecount.head(15)
```

```python
# droping the column consist of all unique ids as it is of no use
valuecount.drop(['index'], axis=1)

# counting total number of frequency days people use the app in one month
v2=valuecount["no_of_days_peopleuse_app"].value_counts()
v2=pd.DataFrame(v2)

# renaming the column to "grouping_people_using_app_acc_to_per day frequency_of_usage"
v2.columns=["grouping_people_using_app_acc_to_per day frequency_of_usage"]
v2.index.name = 'no_of_days'

# calculating the percentage of users using the app according to the frequency of usage
v2["%age of users using app according to per day freq of usage"]= (v2["grouping_people_using_app_acc_to_per day frequency_of_usage"]/v2["grouping_people_using_app_acc_to_per day frequency_of_usage"].sum())*100
v2.sort_index(level=1)

# sorting the values according to the frequencyof daily usage
v3=v2.sort_values(by=['no_of_days'])
v3.set_index([pd.Index(range(1,32))])
v3.head()
```


# Plotting the "POWER USER CURVE"

```python
plt.rcParams["figure.figsize"] = [17,9]
ax = v3[["%age of users using app according to per day freq of usage"]].plot(kind='bar', color=['dodgerblue'], fontsize=13)
ax.legend(loc='upper left');
ax.set_alpha(0.8)
ax.set_title("Power User Curve",fontsize=18)
ax.set_ylabel("percentage of users", fontsize=18);
ax.set_xlabel("Total active days in a month", fontsize=18);
ax.set_yticks([0,10,20,30,40,50,60])
# plotting the values on each bar 
for i in ax.patches:
    ax.text(i.get_x()-.2, i.get_height()+0.05, str(round((i.get_height()), 2))+'%', fontsize=11, color='black', rotation=45)
```
